Task() Develop a python program,whare it checks if the given number is prime or not?

```python
def is_prime(num):
  """
  Checks if a given number is prime.

  Args:
    num: The number to check.

  Returns:
    True if the number is prime, False otherwise.
  """
  if num <= 1:
    return False  # Numbers less than or equal to 1 are not prime
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False  # If the number is divisible by any number from 2 to its square root, it's not prime
  return True  # Otherwise, it's prime

# Example usage:
number_to_check = 17
if is_prime(number_to_check):
  print(f"{number_to_check} is a prime number")
else:
  print(f"{number_to_check} is not a prime number")

number_to_check = 10
if is_prime(number_to_check):
  print(f"{number_to_check} is a prime number")
else:
  print(f"{number_to_check} is not a prime number")
```

`for i in range(2, int(num0.5) + 1):**: This is the core of the prime check. It iterates through numbers starting from 2 up to the square root ofnum` (inclusive). We only need to check up to the square root because if a number has a divisor greater than its square root, it must also have a divisor smaller than its square root.
if num % i == 0:: Inside the loop, it checks if num is divisible by the current number i using the modulo operator (%). If the remainder is 0, it means num is divisible by i, and thus it's not prime. In this case, the function returns False.
return True: If the loop completes without finding any divisors (meaning the number was not divisible by any number from 2 up to its square root), it means the number is prime, and the function returns True.

```python
def analyze_numbers():
  """
  Reads a list of numbers from the user and calculates their mean, minimum, and maximum.
  """
  numbers = []
  while True:
    user_input = input("Enter a number (or 'done' to finish): ")
    if user_input.lower() == 'done':
      break
    try:
      numbers.append(float(user_input))
    except ValueError:
      print("Invalid input. Please enter a number or 'done'.")

  if not numbers:
    print("No numbers were entered.")
    return

  mean = sum(numbers) / len(numbers)
  minimum = min(numbers)
  maximum = max(numbers)

  print(f"Mean: {mean}")
  print(f"Minimum: {minimum}")
  print(f"Maximum: {maximum}")

# Run the function
analyze_numbers()
```

Explaination:
The above code is to help you analyze a set of numbers you provide. It works by first asking you to enter numbers one by one. You can type 'done' when you're finished. The program then takes all the numbers you entered and figures out the average (mean), the smallest number (minimum), and the largest number (maximum) from that set. Finally, it shows you these calculated results.

```python
def is_armstrong_number(num):
  """
  Checks if a given number is an Armstrong number.

  Args:
    num: The number to check.

  Returns:
    True if the number is an Armstrong number, False otherwise.
  """
  # Convert the number to a string to get its digits
  num_str = str(num)
  # Get the number of digits
  n = len(num_str)
  # Calculate the sum of the nth power of each digit
  sum_of_powers = sum(int(digit) ** n for digit in num_str)
  # Check if the sum is equal to the original number
  return sum_of_powers == num

# Get input from the user
try:
  number_to_check = int(input("Enter a number to check if it's an Armstrong number: "))
  if is_armstrong_number(number_to_check):
    print(f"{number_to_check} is an Armstrong number")
  else:
    print(f"{number_to_check} is not an Armstrong number")
except ValueError:
  print("Invalid input. Please enter an integer.")
```

3.It counts how many digits are in the number.
4.For each digit, it raises that digit to the power of the total number of digits.
5.It adds up all these results.
6.Finally, it checks if this sum is equal to the original number you entered. If it is, the number is an Armstrong number; otherwise, it's not.

Prompt 3():Write a Python Function to check wheather the given inputs are prime or palindrome.


```python
def is_prime(num):
  """
  Checks if a given number is prime.

  Args:
    num: The number to check.

  Returns:
    True if the number is prime, False otherwise.
  """
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

def is_palindrome(text):
  """
  Checks if a given string is a palindrome.

  Args:
    text: The string to check.

  Returns:
    True if the string is a palindrome, False otherwise.
  """
  # Remove spaces and convert to lowercase for case-insensitive check
  cleaned_text = "".join(text.split()).lower()
  return cleaned_text == cleaned_text[::-1]

def check_prime_or_palindrome():
  """
  Gets input from the user and checks if it's a prime number or a palindrome based on user choice.
  """
  choice = input("Enter 'prime' to check for primality or 'palindrome' to check for palindrome: ").lower()

  if choice == 'prime':
    try:
      user_input = input("Enter a number to check for primality: ")
      num = int(user_input)
      if is_prime(num):
        print(f"{user_input} is a prime number.")
      else:
        print(f"{user_input} is not a prime number.")
    except ValueError:
      print("Invalid input. Please enter an integer for primality check.")

  elif choice == 'palindrome':
    user_input = input("Enter a word or phrase to check for palindrome: ")
    if is_palindrome(user_input):
      print(f"'{user_input}' is a palindrome.")
    else:
      print(f"'{user_input}' is not a palindrome.")

  else:
    print("Invalid choice. Please enter 'prime' or 'palindrome'.")


# Run the function
check_prime_or_palindrome()
```

Explaination:
The code  check if something you type in is either a prime number or a word/phrase that reads the same forwards and backward (a palindrome).

First, it will ask you if you want to check for "prime" or "palindrome".

1.If you say "prime", it will then ask you to enter a whole number. It will then tell you if that number is prime or not.
2.If you say "palindrome", it will ask you to type a word or phrase. It will then tell you if that word or phrase is a palindrome or not, ignoring spaces and whether you use capital letters or not.
3.If you type something else when asked for "prime" or "palindrome", it will tell you it was an invalid choice.

Promt 4():Write a python function to check the sum of the first N natural numbers (user input).

```python
def sum_of_natural_numbers(n):
  """
  Calculates the sum of the first n natural numbers.

  Args:
    n: The number of natural numbers to sum (a positive integer).

  Returns:
    The sum of the first n natural numbers, or an error message for invalid input.
  """
  if not isinstance(n, int) or n < 1:
    return "Invalid input. Please enter a positive integer."
  else:
    return n * (n + 1) // 2

# Get input from the user
try:
  num_terms = int(input("Enter a positive integer (N) to calculate the sum of the first N natural numbers: "))
  result = sum_of_natural_numbers(num_terms)
  print(f"The sum of the first {num_terms} natural numbers is: {result}")
except ValueError:
  print("Invalid input. Please enter an integer.")
```

This code is designed to calculate the sum of a series of counting numbers (natural numbers) starting from 1, up to a number you choose.

Here's how it works:

1.It will ask you to enter a positive whole number (like 1, 2, 3, and so on). This number is "N".
2.It then uses a quick mathematical trick (a formula) to figure out what you get when you add up all the whole numbers from 1 all the way up to the number you entered (N).
3.Finally, it tells you the total sum. If you don't enter a positive whole number, it will let you know that the input is not valid.

 write a Python program to calculate the sum of

Prompt 5(): Write a python function to calculate the sum of odd numbers and even numbers in a given tuple.

```python
def sum_odd_even_in_tuple():
  """
  Reads a tuple of numbers from the user and calculates the sum of odd and even numbers.
  """
  user_input = input("Enter a tuple of numbers separated by commas (e.g., 1,2,3,4,5): ")
  try:
    # Convert the input string to a tuple of numbers
    numbers_tuple = tuple(float(x.strip()) for x in user_input.split(','))

    sum_odd = 0
    sum_even = 0

    for number in numbers_tuple:
      # Check if the number is an integer before checking for odd/even
      if number.is_integer():
        if int(number) % 2 == 0:
          sum_even += int(number)
        else:
          sum_odd += int(number)
      else:
        print(f"Skipping non-integer value: {number}")


    print(f"Sum of odd numbers: {sum_odd}")
    print(f"Sum of even numbers: {sum_even}")

  except ValueError:
    print("Invalid input. Please enter numbers separated by commas.")
  except Exception as e:
    print(f"An error occurred: {e}")

# Run the function
sum_odd_even_in_tuple()
```

Explaination:
This code is designed to take a list of numbers you give it (like in a tuple) and then separate those numbers into two groups: the odd ones and the even ones. It then adds up all the numbers in the odd group and all the numbers in the even group and tells you the total for each group. If you enter anything that isn't a whole number, it will let you know and skip it.
